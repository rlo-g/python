#### 모듈(라이브러리) 사용하기

```python
import matplotlib.pyplot as plt
```

```python
y = [2, 4, 6]

# plot - 선 그래프(Line plot)
plt.plot(y)    # plot 기준 : y
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

# x, y의 개수가 맞지 않으면 에러 발생!
plt.plot(x, y)
plt.show()
```

```python
# line style (= ls)

plt.plot(x,y, linestyle = '--')  # linesstyle => ls 입력 가능
plt.show()
```

```python
# Marker

plt.plot(x,y, marker='o')
plt.show()
```

```python
x = [2, 4, 6, 8]
y = [10,13, 16, 20]
y2= [10, 20, 15, 12]

plt.plot(x, y, ls='-', marker='o')
plt.plot(x, y2, ls=':')
plt.show()
```

```python
# line width 선 두께 (lw)

plt.plot(x,y, linewidth=6)     # linewidth => lw
plt.show()
```

```python
# line color 선 색상

plt.plot(x,y,color='yellow')
plt.show()
```

```python
# Marker color (mfc)

plt.plot(x,y,color='#91C8E4', marker='D', markerfacecolor='#749BC2')
plt.show()
```

```python
# 마커 선 색상 - markeredgecolor (mec)

plt.plot(x,y,color='#91C8E4', marker='D', markerfacecolor='#749BC2', mec='#F6F4EB')
plt.show()
```

```python
import numpy as np

x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, marker='o', ls='--', color='#4A55A2', ms=12, mfc='#7895CB', mec='#4A55A2', mew=4, linewidth=4)
plt.show()
```

```python
# 그래프 범위 지정

plt.plot(x,y, marker='o', ls='--', color='#9EB384', ms=12, mfc='#CEDEBD', mec='#435334', mew=4, linewidth=4)
plt.xlim(-2,8)      # x축 범위 지정 (최소, 최대)
plt.ylim(-4, 13)    # y축 범위 지정
plt.show()
```

```python
# 격자무늬 표현
#  - 틱의 위치를 보기 위해서
#  - tick : 그래프 축의 위치

plt.plot(x,y, marker='o', ls='--', color='#4A55A2', ms=12, mfc='#7895CB', mec='#4A55A2', mew=4, linewidth=4)

plt.xticks([0,3,6])    # x축 틱 설정 >> 설정 안 하면 자동으로 설정 됨
plt.yticks([1,2,7])
plt.grid()
plt.show()
```

```python
# 여러 plot 그리기

x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

# 그래프에 이름 부여
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')

# 범례
plt.legend()

plt.show()
```

```python
plt.plot(x, label='엑스')
plt.plot(y, label='와이')
plt.plot(z, label='제트')

# 축에 이름 표시
plt.xlabel('x축')
plt.ylabel('y축', rotation=0)

# 그래프의 이름
plt.title('x, y, z 그래프')

plt.legend()
plt.show()
```
