```python
# userId, itemId, rating, timestamp

ratings = np.loadtxt('ratings.txt', delimiter=',', dtype=np.int64)
ratings
```

```python
print(ratings.ndim)
print(ratings.shape)
print(ratings.size)
```

```python
#    

ratings[:,2].mean()
```

```python
# userId 1 user   

ratings[:, 0]              ## ratings 0  
id = ratings[:, 0] == 1   ## 0    1   (True)

ratings[id, 2].mean()     ## id 1  2 
```

```python
# userId 5 user   

ratings[ratings[:, 0] == 5, 2].mean()
```

```python
#       

print(np.unique(ratings[:, 0]))
print(np.unique(ratings[:, 0]).shape)

print(ratings[:,0].shape)


# user    

unique_id = np.unique(ratings[:, 0])

li = []

for user_id in unique_id:
    user_mean = ratings[ ratings[:, 0] == user_id , 2 ].mean()  ##  : boolean  , : 2 
    print(user_mean)
    li.append([user_id, user_mean])  ##   userId  user mean 
```

```python
li[:5]
```

```python
# itemId    

unique_itemId= np.unique(ratings[:, 1])

item_li = []

for item_id in unique_itemId:
    itemId_mean = ratings[ratings[:,1]==item_id, 2].mean()   ##  item_id  (True) 2  
    item_li.append([item_id, itemId_mean])
```

```python
item_li[:5]
```
