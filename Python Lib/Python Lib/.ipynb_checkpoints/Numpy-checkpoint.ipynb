### 1차원 배열(ndarray) 만들기

```python
list1 = [1, 2, 3, 4, 5]

# np의 배열을 생성하는 함수
arr = np.array(list1)  # list1을 가지고 배열 생성
arr
```

### 2차원 배열(array) 만들기

```python
list2 = [[1, 2, 3], [4, 5, 6]]  ## 2차원 리스트 생성 (대괄호 수 = 차원)
arr2 = np.array(list2)          ## 2차원 리스트로 2차원 배열 생성
arr2
```

### 인덱싱, 슬라이싱

#### - 인덱싱

```python
arr = np.array([[1,2,3], [4,5,6]])
arr
```

```python
# 행 인덱싱

print(arr[0])     # 리스트 안에 있는 첫번째 요소
```

```python
# arr에서 3 인덱싱

print(arr[0][2])  # 첫번째 요소 안에 있는 세번째 요소
print(arr[0, 2])  # - 0행 2열
```

#### - 슬라이싱

```python
# np.arange(시작, 끝, 증감)
#  - 디폴트 : (0, 작성값, 1)

arr1 = np.arange(10) 
arr1
```

```python
np.arange(2, 10, 2)
```

```python
## 슬라이싱

print(arr1)
print(arr1[2:6])
```

```python
# 2차원 범위 배열 생성
# reshape : 배열의 크기를 다시 지절

arr2 = np.arange(50).reshape(5, 10)
arr2
```

```python
# 2차원 배열 슬라이싱
# [행의 범위 , 열의 범위]

arr2[ :2 , : ]  # :n (처음부터 n-1), : (전체)
```

```python
arr2[ : , 0]   # [슬라이싱, 인덱싱]
```

```python
arr2[ : , 4:8]
```

```python
arr2[ :4, :3]
```

```python
arr2[2, 1]
```

```python
# 21, 32 한 번에 가져오기

arr2[ [2,3] , [1,2] ]  # [ [행1, 행2] , [열1, 열2]]  >> [행렬1, 행렬2]
```

### 영화 평점 데이터 연산

```python
# userId, itemId, rating, timestamp

ratings = np.loadtxt('ratings.txt', delimiter=',', dtype=np.int64)
ratings
```

```python
print(ratings.ndim)
print(ratings.shape)
print(ratings.size)
```

```python
# 전체 평점 평균 구하기

ratings[:,2].mean()
```

```python
# userId가 1인 user의 평점 평균 구하기

ratings[:, 0]              ## ratings의 0번째 열만 가져오기
id = ratings[:, 0] == 1   ## 0번째 열 중 값이 1인 것만 가져오기 (True)

ratings[id, 2].mean()     ## id가 1인 값들 2열의 평균값
```

```python
# userId가 5인 user의 평점 평균 구하기

ratings[ratings[:, 0] == 5, 2].mean()
```

```python
# 고유한 요소 한 개씩만 남겨서 배열 생성

print(np.unique(ratings[:, 0]))
print(np.unique(ratings[:, 0]).shape)

print(ratings[:,0].shape)


# user 별 평정 평균 구하기

unique_id = np.unique(ratings[:, 0])

li = []

for user_id in unique_id:
    user_mean = ratings[ ratings[:, 0] == user_id , 2 ].mean()  ## 행 : boolean 타입으로 색인, 열: 2번째 값
    print(user_mean)
    li.append([user_id, user_mean])  ## 빈 리스트에 userId 별 user mean 저장
```

```python
li[:5]
```

```python
# itemId 별 평점 평균 구하기

unique_itemId= np.unique(ratings[:, 1])

item_li = []

for item_id in unique_itemId:
    itemId_mean = ratings[ratings[:,1]==item_id, 2].mean()   ## 행이 item_id와 같을 때(True) 2열 값의 평균
    item_li.append([item_id, itemId_mean])
```
