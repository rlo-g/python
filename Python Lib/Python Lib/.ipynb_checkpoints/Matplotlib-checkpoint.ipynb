```python
import matplotlib.pyplot as plt
```

```python
y = [2, 4, 6]

# plot -  (Line plot)
plt.plot(y)    # plot  : y
plt.show()
```

```python
x = [1, 2, 3]
y = [2, 4, 6]

# x, y     !
plt.plot(x, y)
plt.show()
```

```python
# line style (= ls)

plt.plot(x,y, linestyle = '--')  # linesstyle => ls  
plt.show()
```

```python
# Marker

plt.plot(x,y, marker='o')
plt.show()
```

```python
x = [2, 4, 6, 8]
y = [10,13, 16, 20]
y2= [10, 20, 15, 12]

plt.plot(x, y, ls='-', marker='o')
plt.plot(x, y2, ls=':')
plt.show()
```

```python
# line width   (lw)

plt.plot(x,y, linewidth=6)     # linewidth => lw
plt.show()
```

```python
# line color  

plt.plot(x,y,color='yellow')
plt.show()
```

```python
# Marker color (mfc)

plt.plot(x,y,color='#91C8E4', marker='D', markerfacecolor='#749BC2')
plt.show()
```

```python
#    - markeredgecolor (mec)

plt.plot(x,y,color='#91C8E4', marker='D', markerfacecolor='#749BC2', mec='#F6F4EB')
plt.show()
```

```python
import numpy as np

x = np.arange(7)
y = [1,4,5,8,9,5,3]

plt.plot(x,y, marker='o', ls='--', color='#4A55A2', ms=12, mfc='#7895CB', mec='#4A55A2', mew=4, linewidth=4)
plt.show()
```

```python
#   

plt.plot(x,y, marker='o', ls='--', color='#9EB384', ms=12, mfc='#CEDEBD', mec='#435334', mew=4, linewidth=4)
plt.xlim(-2,8)      # x   (, )
plt.ylim(-4, 13)    # y  
plt.show()
```

```python
#  
#  -    
#  - tick :   

plt.plot(x,y, marker='o', ls='--', color='#4A55A2', ms=12, mfc='#7895CB', mec='#4A55A2', mew=4, linewidth=4)

plt.xticks([0,3,6])    # x   >>      
plt.yticks([1,2,7])
plt.grid()
plt.show()
```

```python
#  plot 

x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

#   
plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')

# 
plt.legend()

plt.show()
```

```python
plt.plot(x, label='')
plt.plot(y, label='')
plt.plot(z, label='')

#   
plt.xlabel('x')
plt.ylabel('y', rotation=0)

#  
plt.title('x, y, z ')

plt.legend()
plt.show()
```
