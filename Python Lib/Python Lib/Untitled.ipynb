```python
import pandas as pd
```

```python
data15 = pd.read_csv('2015.csv', encoding='euc-kr', index_col='관서명')
data16 = pd.read_csv('2016.csv', encoding='euc-kr', index_col='관서명')
data17 = pd.read_csv('2017.csv', encoding='euc-kr', index_col='관서명')
```

```python
df15 = pd.DataFrame(data15)
df15
```

```python
occur15 = df15[df15['구분']=='발생건수']

occur15
```

```python
total15 = occur15.loc[:,'살인':'폭력'].sum(axis=1)
total15
```

```python
df16 = pd.DataFrame(data16)
df16
```

```python
occur16 = df16[df16['구분']=='발생건수']

total16 = pd.Series(occur16.loc[:,'살인':'폭력'].sum(axis=1), name='2016총계')

total16
```

```python
rate1516 = (total16-total15)/total15*100
#(total16-total15)/total15*100

rate1516
```

```python
df17 = pd.DataFrame(data17)
df17
```

```python
occur17 = df17[df17['구분']=='발생건수']

total17 = occur17.loc[:,'살인':'폭력'].sum(axis=1)
total17.drop('광주지방경찰청', inplace=True)
total17
```

```python
rate1617 = (total17-total16)/total16*100
rate1617
```

```python
result = pd.concat([total15, rate1516, total16, rate1617, total17], axis=1)  # 컬럼을 옆으로 채우므로 axis=1
result
```

```python
# 컬럼명 수정1

result.columns = ['2015총계', '2015-2016증감율', '16총계', '2016-2017증감율', '17총계']
```

```python
# 컬럼명 수정2
# - Series에 이름(name) 주기

total16.name = '16총계'
```

```python
# 컬럼명 수정3

result.rename(columns={'2015총계':'2015합계', '16총계':'16합계'}, inplace=True)  # inplace=True  >> 해야 반영됨
```
