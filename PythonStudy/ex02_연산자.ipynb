## 연산자 종류



```python
# 나누기, 나머지, 나누기(몫) 구하기

num1 = 10
num2 = 7

print(num1/num2)
print(num1//num2)
print(num1%num2)
```

```python
# 문자열끼리 더하기 연산 가능!
str1 ="7"
str2 = "10"
print(str1+str2)
```

```python
# 문자열과 숫자는 더하기 연산 불가능
num1 = 10
str2 = "7"
# print(num1+str2)
```

```python
# Java Scanner처럼 키보드로 데이터 입력받기
# input()
# int('문자열')
# float('문자열')
# str(데이터)
# type(변수) -> 데이터 타입 확인 가능

num = input("정수 입력 >> ")
print(int(num)+1)
```

### 실습



```python
num1 = int(input("정수를 입력하세요 >> "))
num2 = int(input("정수를 입력하세요 >> "))

print("더하기 결과 : %d" % (num1+num2))
print("빼기 결과 : {}".format(num1-num2))
print(f"곱하기 결과 : {num1*num2}")
print(f"나누기 결과 : {num1/num2}")
```

### 실습


```python
time = int(input("시간 입력 >> "))
hour = time // 3600
minute = time%3600//60
second = time%60

print("{}시간 {}분 {}초".format(hour, minute, second))
```

## 복합, 비교 연산자





```python
a = 3
b = 7
print(a > b)
print(a == b)
print(a != b)
```

## 논리연산자
   - Java : &&, ||, !
   - Python : and, or, not
   


```python
a = 3
b = 8

print(not a < b)

print(a > b and 10 == 20)  ## false and false = false
print(True and True)
print(a < b and 10!=20)

print(a < b or 10==20)    ## true or false = true
print(True or False)
```

## 삼항 연산자




```python
score = 80
result = '합격' if score >= 60 else '불합격'

print(result)
```

### 실습



```python
num = int(input("정수 입력 >> "))

result = '짝수' if num % 2 == 0 else '홀수'

print(f'{num}은(는) {result}입니다.')
```
