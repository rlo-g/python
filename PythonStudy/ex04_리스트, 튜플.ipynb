## 리스트



```python
# 리스트 생성

a = []  # 타입 :list
b = [1, 2, 3]
c = ['My', 'Name']
d = [1, 2, 'A', 'B']          ## 숫자와 문자 같이 입력 가능
e = [1, 2, 'a', [3, 4, 'b']]  ## list 안에 list 구조도 가능


print(type(a))
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
# 인덱싱(indexing)
# 무엇인가를 '가리킨다'
# 리스트명[인덱스] :  인덱스에 위치한 값을 반환

# 슬라이싱(slicing)
# 무엇인가를 '잘라낸다'
# 리스트명[ 시작 인덱스 : 종료 인덱스] : 시작 인덱스부터 종료 '인덱스 전'까지의 값을 반환
```

```python
li1 = [2, 3, 5, 7, 1]
print(li1[2])
print(li1[0])
print(li1[2] + li1[-1])
```

```python
li2 = [1, 2, 3, ['a', 'b', 'c']]

# temp변수에 ['a', 'b', 'c']를 저장하고 출력
temp = li2[3]

print(temp)

# li2에서 문자 'b'만 인덱싱해서 출력
print(li2[3][1])  # li2[3]에서 [1]번째 요소를 가져온다
```

```python
li3 = [1, 4, 6, 7, 2]
print(li3[1:3])
print(li3[:3])  # 처음부터 3번째 전까지
print(li3[1:])  # 1번째부터 마지막까지
```

```python
# 리스트 연산

li4 = [1, 2, 3]
li5 = [3, 4, 5]

print(li4*3)      # 문자열처럼 리스트 반복
print(li4 + li5)  # 리스트가 합쳐짐
```

```python
# 리스트 값 추가하기
# 리스트명.append(값)

li6 = [1, 2, 3]
li6.append(8)   # 추가한 값은 맨마지막 순서로 들어감
print(li6)
```

```python
# 원하는 위치에 리스트 값 삽입
# 리스트명.insert(인덱스, 값)

li7 = [0, 1, 2, 3, 4]
li7.insert(1, 5)
print(li7)
```

```python
# 리스트 값 수정
# 1. 리스트명[인덱스] = 값
# 2. 리스트명[슬라이싱] = [값]

li8 = [0,1, 2, 3, 4]
li8[1] = 7
print(li8)

li8[2:4] = [9]
print(li8)
```

```python
array = [1, 2, 3, 4, 5]

# 숫자 5를 7로 변경하기
array[-1] = 7

# 숫자 2를 리스트['a', 'b', 'c']로 변경
array[1] = ['a', 'b', 'c']

# 4, 7을 'd', 'e', 'f', 'g'로 변경
array[3:] = 'd', 'e', 'f', 'g'

# 1을 'h', 'i', 'j'로 변경
array[0:1] = 'h', 'i', 'j'  ## 슬라이싱으로 데이터에 접근해야 각각의 요소를 넣을 수 있다

print(array)
```

```python
# 리스트 값 삭제
# 1. 인덱스를 이용하는 방법
# del 리스트명[인덱스]

# 2. 값을 이용하는 방법
# 리스트명.remove(값)

li9 = ['a', 'c', 'b', 'c', 'd', 'e']
del li9[1]
print(li9)     # 맨 앞에 있는 c만 삭제

li9.remove('c')
print(li9)

# li9.remove('f')  >> 에러발생!
```

```python
# 리스트 정렬
# 리스트명.sort()  >  기본적으로 오름차순 정렬

li10 = [3, 6, 7, 2, 4, 1, 7]
li10.sort()
print(li10)

# 내림차순 정렬
li10.reverse()
print(li10)
```

```python
# 리스트 순서 뒤집기
# 리스트명.reverse()

li11 = [3, 6, 7, 2, 4, 1, 7]
li11.reverse()
print(li11)

li11.sort(reverse = True)  # sort()는 기본적으로 내림차순이 False로 적용되어 있다 > True로 하면 내림차순 정렬이 된다
print(li11)                # 함수 정보 확인 단축키 : shift + tab
```

```python
# 리스트 특정 값의 위치 반환(찾기)
# 리스트명.index(값)

li12 = ['a', 'b', 'c', 'd', 'c',  5, 10]
print(li12.index('c'))
print(li12.index(5))
```

```python
# 리스트 값의 길이(개수) 확인

print(len(li12))
print(len('안녕하세요'))
```

## 튜플



```python
# 튜플 생성
a = ()
b = (1, 2, 3)
c = ('a', 'b', 'c', (1, 2, 3))
d = 1, 2, 3, 4, 5
a, b = (1,2,3), 4
name = ('Hammi', )  # 데이터 한 개를 튜플에 담을 때는 뒤에 ,를 찍어준다

print(a)
print(type(a))
print(b)
print(c)
print(d)
print(type(d))
print(a,b)
print(name)
print(type(name))  # 뒤에 ,가 없이 한 개의 데이터가 들어있다면 타입은 String
```

```python
# 튜플 인덱싱/슬라이싱/길이

tu1 = (0, 1, 2, 3, ('a','b'), 5)
print(tu1[3])
print(tu1[1:5])
print(len(tu1))

## 튜플은 추가, 수정, 삭제 불가능
```

## 멤버연산자



```python
str1 = '파이썬 최고'
print('파이썬' in str1)  # '파이썬'이 str1에 들어있는가?
print('파이썬' not in str1)  # '파이썬'이 str1에 들어있지 않은가?

li = [23, 44, 11]
print(2 in li)
print(20 not in li)

if '파이썬' in str1:
    print('포함')
else:
    print('미포함')
```

### 실습



```python
lst = ['딸기', '바나나', '수박', '체리', '포도']

print('*'*50)
word = input('검색할 문자를 입력하세요 >> ')

if word in lst:
    print(f'{word}는 리스트 {lst.index(word)}번째 인덱스에 들어있습니다.')
else:
    print(f'{word}는 리스트에 들어있지 않습니다.')
print('*'*50)
```

```python
s = "Hi, My Name Is Eunji"

print('*'*50)
word = input('검색할 문자를 입력하세요 >> ')

if word in s:
    print(f'{word}는 {s.count(word)}번 들어있습니다.')
else:
    print(f'{word}는 문자열에 들어있지 않습니다.')
print('*'*50)
```
