# 조건문

## - if문



```python
if True:
    print('실행문1')
print('조건문 종료')
print()

if False:
    print('실행문1')
    print('실행문2')
print('조건문 종료')
print()

if True:
    print('실행문a')
    if True:
        print('실행문b')
print('조건문 종료!')
print()

if True:
    print('실행문1')
    if False:
        print('실행문2')
    print('실행문3')
print('조건문 종료...')
```

## - if else문
 - True일 경우 if라면, False일 경우 else로 조건문 실행






## - elif문
 - Java의 else if와 같음
 
 
 

### 실습



```python
score = int(input('점수 입력 >> '))

if score >= 90:
    print(f'{score}점은 A학점 입니다.')
elif score >= 80:
    print(f'{score}점은 B학점 입니다.')
elif score >= 70:
    print(f'{score}점은 C학점 입니다.')
elif score >= 60:
    print(f'{score}점은 D학점 입니다.')
else:
    print(f'{score}점은 F학점 입니다.')
    
```

### 실습





```python
print('----- [SM Coffee] ----')

money = int(input('금액을 입력하세요 >> '))

print('커피를 고르세요.')
select = int(input('[1]에스프레소(600원) [2]아메리카노(1000원) [3]카페라떼(1300원) >> '))

if select == 1:
    price = 600     ## print(money - 금액)이 동일하므로 밑으로 빼고 price를 변수로 설정
elif select == 2:
    price = 1000
elif select == 3:
    price = 1300

if money < price:
    print('돈이 부족해요...T^T')
    price = 0   ## 돈이 부족하면 넣은 돈을 그대로 돌려줘야 하므로 money - 0
    
print(f'거스름돈 : {money - price}원')
        
        
```
