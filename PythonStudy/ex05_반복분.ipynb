## while문



```python

while True:
    num1 = int(input('첫 번째 정수 입력 >> '))
    num2 = int(input('두 번째 정수 입력 >> '))
    
    if num1 == 0 and num2 == 0:
        print('프로그램이 종료되었습니다')
        break
    print('두 정수의 합 : ', num1+num2)
```

```python
num1 = 1
num2 = 1

while num1 != 0 or num2 != 0:
    num1 = int(input('첫 번째 정수 입력 >> '))
    num2 = int(input('두 번째 정수 입력 >> '))
    if num1 != 0 or num2 != 0:
        print('두 정수의 합 : ', num1+num2)
print('프로그램이 종료되었습니다')
```

```python
import random  ##랜덤함수 불러오기

ans = random.randint(1, 50) 

while True:
    num = int(input('숫자를 입력하세요 >> '))
    
    if num > ans:
        print(f'{num}보다 작은 수 입니다.')
    elif num < ans:
        print(f'{num}보다 큰 수 입니다.')
    else:
        print('정답!!!')
        break
    
```

```python
score_list = [90, 45, 70, 60, 55, 90]

cnt = 1 #반복횟수
for score in score_list:
    if score >= 60:
        print(f"{cnt}번 학생은 합격입니다.")
    else:
        print(f"{cnt}번 학생은 불합격입니다")
    cnt += 1
        
 ## 리스트 안에 동일한 숫자가 있을 경우, .index() 사용 시 중복 인덱스로 출력됨... > 반복횟수 카운팅 변수 만들기 : cnt
```

```python
list1 = [[1,2], [3,4], [5,6]]
print(len(list1))    #리스트 길이 : 3
for li in list1:
    print(li)        ## list 안에 들어있는 li 요소의 데이터 타입 : list
    for i in li:
        print(i)     ## li 안에 들어있는 리스트의 요소들 차례대로 출력
    print('-------')
```

```python
list2 = [ 4, 5, 2, 1, 99, 15, 2, 7, 27]

max = list2[0]   ## 리스트 안에 있는 값을 활용하여 값 비교하기
for i in list2:
    if i > max:
        max = i
print(max)
```

```python
list2 = [-5, -7, -23, -44]

min = list2[0]

for i in range(len(list2)):
    if min > list2[i]:
        min = list2[i]
print(min)
    
```

#### 리뷰 길이의 평균값


```python
total = 0
for i in range(len(movie_review)):
    total += len(movie_review[i][0])
print(f"리뷰의 평균 길이는 '{total/len(movie_review)}'입니다.")
```

```python
total = 0
for txt, pn in movie_review:   ## a, b = [1, 2]
    total += len(txt)
print(f"리뷰의 평균 길이는 '{total/len(movie_review)}'입니다.") 
```

#### 긍정/부정 리뷰 개수 세기



```python
p = 0
n = 0

for text, pn in movie_review:
    if pn == 1:
        p += 1
    elif pn == 0:
        n += 1
print(f"긍정 '{p}'개")
print(f"부정 '{n}'개")
```

```python
total = 0
for txt, pn in movie_review:
    total += pn        ## 부정은 0이므로 총 더한 값은 긍정의 개수와 같음
    
print(f"긍정 '{total}'개")
print(f"부정 '{len(movie_review)-total}'개")
```

```python
pnLi = []
for txt, pn in movie_review:
    pnLi.append(pn)
    
print(f"긍정 '{pnLi.count(1)}'개")
print(f"부정 '{pnLi.count(0)}'개")
```

#### 리뷰 최대 길이/ 리뷰 내용 출력하기



```python
max = ''

for txt, pn in movie_review:
    if len(txt) > len(max):
        max = txt

print(f"리뷰 최대 길이는 '{len(max)}'입니다.")
print(max)
```

```python
max = len(movie_review[0][0])

for i in range(len(movie_review)):
    if max < len(movie_review[i][0]):
        max = len(movie_review[i][0])
        cnt = movie_review[i][0]
    
print(f"리뷰 최대 길이는 '{max}'입니다.")
print(cnt)
```

#### 검색 데이터가 포함된 리뷰 개수


```python
search = input('검색할 데이터 입력 >> ')

cnt = 0

for txt, pn in movie_review:
    if search in txt:
        cnt += 1
print(f"'{search}'가 포함된 리뷰는 {cnt}개 입니다.")
```
