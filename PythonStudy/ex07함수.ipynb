```python
def add(*args):
    sum = 0
    for i in args:
        sum += i
    return sum
```

```python
print(add(2,3,4))
```

#### 야구게임



```python
li1 = []

def isDuplicate(li1):
    for i in range(0,len(li1)-1):
        if li1[i] == li1[i+1]:
            check = 'False'
            break
        else:
            check = 'True'
    return check
```

```python
li1 = [2, 7, 5, 3, 3]
check = isDuplicate(li1)
print(check)
```
